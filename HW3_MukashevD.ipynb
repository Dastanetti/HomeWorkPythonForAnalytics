### Task 1.1 Создать новый признак price_per_sq_lot, который будет содержать среднюю стоимость за один кв. метр общей площади

```python
df['1price_per_sq_lot']=df['price']/df['sqft_lot']
```

```python
df[['price','sqft_lot','1price_per_sq_lot']].sort_values('sqft_lot').head(10)
```

```python
df['sqft_lot'].nunique()
```

```python
df['sqft_lot'].value_counts()
```

```python
df['1price_per_sq_lot'].values
```

```python
price_per_sq_lot = df.groupby('sqft_lot').agg({'1price_per_sq_lot' : 'mean'})
price_per_sq_lot[0:1].values
```

```python
price_per_sq_lot.head(10)
```

```python
age_days = []
value_squ = 0
for i, row in df.iterrows():
    if row['sqft_lot'] in price_per_sq_lot.index:
        value_squ = age_days.append(price_per_sq_lot[i:i+1].values)
    else:
        age_days.append(value_squ)

age_days[0:10]
```

```python
df['price_per_sq_lot'] = age_days
df.head()
```

### Task 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома

```python
df['delta_renovated'] = 0
```

```python
df.loc[df['yr_renovated'] != 0, 'delta_renovated'] = df['yr_renovated'] - df['yr_built']
```

```python
df.delta_renovated.sort_values()
```

### Task 1.3 Создайте признаки года продажи, месяца продажи

```python
df['date'] = pd.to_datetime(df['date'])
df['year_sale'] = df['date'].apply(lambda x: x.date().year)
df['month_sale'] = df['date'].apply(lambda x: x.date().month)
d = dict(enumerate(calendar.month_abbr))
df['month_sale'] = df['month_sale'].map(d)
df.sample(5)
```

### Task1.4 Удалите признаки date, zipcode, lat, long


```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
df.head()
```
